```q
loc:(raze data)?"S"
cnt: count data
start_x:"i"$loc%cnt
start_y:"i"$mod[loc;cnt]
start_x, start_y
```

```q
step_map:([] prv:"UUUDDDRRRLLL";ch:"|7F|JL-J7-LF";nxt:"ULRDLRRUDLUD");
step_map:("UUUDDDRRRLLL";"|7F|JL-J7-LF";"ULRDLRRUDLUD");
step_map
```

```q
step_op:(("U";-1;0);("D";1;0);("L";0;-1);("R";0;1))
step_op
```

```q
step_count:(cnt, cnt) #-1;
.tmp.far:0;
step_count[start_x;start_y]:0;

next_step:{[last_step;ch]
    step_map[2;where (last_step in 'step_map[0]) & (ch in 'step_map[1])]
    };

process_ele:{

    if[3 > count x;
        /show "Empty Prev";
        :()
        ];

    new_loc:(x[0],x[1])+(1_step_op[step_op[;0]?x[2]]);
    ch:data[new_loc[0];new_loc[1]];

    if[ch in ". S";
        :()
        ];

    if[step_count[new_loc[0];new_loc[1]] = -1;
        step_count[new_loc[0];new_loc[1]]: (1 + step_count[x[0];x[1]]);
        .tmp.far: step_count[new_loc[0];new_loc[1]];
        :(new_loc, next_step[x[2];ch])
        ];

    if[ .tmp.far <= ((1 + step_count[x[0];x[1]])+step_count[new_loc[0];new_loc[1]])%2;
        .tmp.far: ((1 + step_count[x[0];x[1]])+step_count[new_loc[0];new_loc[1]])%2;
        ];

    show .tmp.far;
    show ((1 + step_count[x[0];x[1]])+step_count[new_loc[0];new_loc[1]])%2;

    :()
    };

steps:((start_x,start_y,"U");
    (start_x,start_y,"D");
    (start_x,start_y,"L");
    (start_x,start_y,"R"));

while[0 < count steps;
    /show "Processing"; show steps;
    steps: ((process_ele ') steps) except enlist ()
    ]

step_count
.tmp.far
```
